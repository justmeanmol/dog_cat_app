```python
import zipfile
zip_ref = zipfile.ZipFile('/content/cat-or-dog-image-classification.zip', 'r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense,Flatten



```

```python
conv_base = VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(150,150, 3)
)
```

```python
conv_base.summary()
```

```python

model = Sequential()

model.add(conv_base)  # pretrained model (VGG16)
model.add(Flatten())

model.add(Dense(256, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.summary()
```

```python
conv_base.trainable=False
```

```python
train_ds = keras.utils.image_dataset_from_directory(
    directory="/content/Train",
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(150,150)
)


val_ds = keras.utils.image_dataset_from_directory(
    directory="/content/Test",
    labels="inferred",
    label_mode="int",
    batch_size=32,
    image_size=(150,150)
)
```

```python
#normalize
def process(image,label):
  image = tensorflow.cast(image/255. ,tensorflow.float32)
  return image,label

train_ds = train_ds.map(process)
val_ds = val_ds.map(process)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(train_ds,epochs=10, validation_data=val_ds,)
```

```python
conv_base.trainable = True

for layer in conv_base.layers[:15]:
    layer.trainable = False


model.compile(
    optimizer=keras.optimizers.Adam(1e-5),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

history_finetune = model.fit(train_ds, validation_data=val_ds, epochs=10)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img_path = "/content/download (4).jpeg"

img = image.load_img(img_path, target_size=(150, 150))
x = image.img_to_array(img) / 255.0
x = np.expand_dims(x, axis=0)

prediction = model.predict(x)[0][0]

if prediction > 0.5:
    print("Dog 🐶")
else:
    print("Cat 🐱")
```
